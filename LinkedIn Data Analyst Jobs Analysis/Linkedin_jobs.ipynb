```python
pip install wordcloud
```

```python
text = ' '.join(df['Job Description'])
tokens = word_tokenize(text) # Tokenize the text into words
stop_words = set(stopwords.words('english')) #create a set of common words in english
filtered_tokens = [word for word in tokens if word.lower() not in stop_words] #remove stopwords from list of tokens
freq_dist = FreqDist(filtered_tokens) #frequency distribution of Filtered tokens

#WORDCLOUD of keywords in Job Description
from wordcloud import WordCloud
wordcloud = WordCloud(width=1200, height=600, max_words=200, background_color='white').generate_from_frequencies(freq_dist)
# Display the generated word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

###### Top 10 keywords in Job Description

```python
# Display the top 10 keywords and their frequencies
print("Top 10 keywords and their frequencies:")
for word, frequency in freq_dist.most_common(20):
    print(f"{word}: {frequency}")
```

## Job Level - Average Salary And Number of Jobs

```python
# Group by 'Level' and calculate mean salary and count of job titles
level_salary_jobs = df_yearly.groupby('Level').agg({'Avg Salary': 'mean', 'title': 'count'}).sort_values(by='Avg Salary', ascending=True)
# Reverse the order of levels
level_salary_jobs = level_salary_jobs[::-1]
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 3))
# Plot average salary with reversed order
sns.barplot(x='Avg Salary', y=level_salary_jobs.index, data=level_salary_jobs, color='indigo', ax=ax1)
ax1.set_xlabel('Average Salary')
ax1.set_ylabel('Level')
ax1.set_title('Average Salary by Level')
for index, value in enumerate(level_salary_jobs['Avg Salary']):
    ax1.text(value, index, f'{value:.2f}', ha='left', va='center')

# Plot number of jobs with reversed order
ax2.barh(level_salary_jobs.index, level_salary_jobs['title'], color='orange')
ax2.set_xlabel('Number of Jobs')
ax2.set_ylabel('Level')
ax2.set_title('Number of Jobs by Level')
ax2.invert_yaxis()  # Reverse the y-axis order
for index, value in enumerate(level_salary_jobs['title']):
    ax2.text(value, index, str(value), ha='center', va='center')
#Remove borderes from chart
for ax in [ax1, ax2]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
plt.tight_layout()
plt.show()
```

###### 74% of the Jobs are available for Mid-Senior-level follwed by Associate  and Entry-Level . As Obvious Associate Jobs are paid higher Salary than Mid-senior-level position followed by Entry-level jobs
