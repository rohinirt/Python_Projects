### Coapplicant Income vs Applicant Income

```python
sns.scatterplot(x='Applicant_Income',y='Coapplicant_Income',hue = 'Status', data = df)
plt.xlabel("Applicant Income")
plt.ylabel("Coapplicant Income")
plt.show()
```

### Loan Amount vs Term 

```python
sns.scatterplot(x='Term',y='Loan_Amount',hue = 'Status', data = df)
plt.xlabel("Term")
plt.ylabel("Loan Amount")
plt.show()
```

### Credit History vs Applicant Income 

```python
sns.scatterplot(x='Applicant_Income',y='Credit_History',hue = 'Status', data = df)
plt.xlabel("Applicant Income")
plt.ylabel(" Credit History")
plt.show()
```

### Gender

```python
ax= sns.countplot(x='Gender',hue = 'Status',data=df)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```
